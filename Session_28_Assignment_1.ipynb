### Problem Statement
* In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
paragraph = soup.get_text(strip=True)
print (paragraph)
```

```python
##Tokenizing sentence
words=nltk.word_tokenize(paragraph)
```

```python
wordFreq=nltk.FreqDist(words)
```

```python
len(words)
```

```python
## Printing the words
for key,value in wordFreq.items():
    print(str(key) + ":" + str(value))
```

```python
##Plotting the top 30 frequncy words
wordFreq.plot(30, cumulative=False)
```

```python
## Removing the stopwords from the wordlist
from nltk.corpus import stopwords
import string
words_no_stopwords =[]
for i in range(len(words)):
    word = nltk.word_tokenize(words[i])
    
    for j in word:
        
        if (j not in stopwords.words('english'))  & (j not in string.punctuation):
         
            #print(j)
            words_no_stopwords.append(j)
```

```python
len(words_no_stopwords)
```

```python
## Plotting top 30 words after removing stop words
freq = nltk.FreqDist(words_no_stopwords)
freq.plot(30,cumulative=False)
```
